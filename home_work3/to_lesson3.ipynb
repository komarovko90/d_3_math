### 3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
1.	окружности,
2.	эллипса,
3.	гиперболы.

```python
from pylab import rcParams
rcParams["figure.figsize"] = 4, 4

R = 5
х0 = 1
y0 = 1
x=[]
y=[]
y2=[]
for i in np.linspace(-4,6, 100):
    x.append(i)
    y.append(np.sqrt(R*R - pow(i-х0, 2))+y0)
    y2.append(-np.sqrt(R*R - pow(i-х0, 2))+y0)

plt.axis([-6,6,-6,6])    
plt.plot(x,y)
plt.plot(x,y2)
plt.grid()
plt.show()
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5.5, 0.5)
Y = np.arange(-5, 5.5, 0.5)
X, Y= np.meshgrid(X, Y)
Z = 2*X + 2*Y
Z2 = 2*X + 2*Y -10
ax.scatter(0, 0, 0, 'z', 50, 'red')
ax.plot_wireframe(X, Y, Z)
ax.plot_wireframe(X, Y, Z2)

show()
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5.5, 0.5)
Y = np.arange(-5, 5.5, 0.5)
X, Y= np.meshgrid(X, Y)
Z = (X**2 + Y**2)
Z2 = -(X**2 + Y**2) 
ax.scatter(0, 0, 0, 'z', 50, 'red')
ax.plot_wireframe(X, Y, Z)
ax.plot_wireframe(X, Y, Z2)

show()
```

### 1. Задание (в программе)
Нарисуйте график функции:
y(x) = k∙cos(x – a) + b
для некоторых (2-3 различных) значений параметров k, a, b

```python
rcParams["figure.figsize"] = 7, 4

x=[]
y=[]
y2=[]
y3=[]
for i in np.linspace(-2*np.pi, 3*np.pi, 200):
    x.append(i)
    y.append(2*np.cos(i-1) + 1)
    y2.append(3*np.cos(i-1) + 2)
    y3.append(0.5*np.cos(i-3) + 3)

# plt.axis([-6,6,-6,6])    
plt.plot(x,y)
plt.plot(x,y2)
plt.plot(x,y3)
plt.grid()
plt.show()
```
